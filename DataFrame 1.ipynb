# Drop Rows with Null Values

```python
nba= pd.read_csv("nba.csv")
nba.tail() #last rows has NAN
```

```python
nba.dropna()
nba.dropna(how = "any")
#by default any row having any missing row in any column
#meaning even if a row has atleast 1 nan then it will remove the entire row 
```

```python
#to target row containing nan in all columns
nba.dropna(how= "all")
```

```python
#to target nan in a certain column
nba.dropna(subset = ["College"])
```

```python
#in either column
nba.dropna(subset = ["College","Salary"])
```
